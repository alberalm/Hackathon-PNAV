```python
import pandas as pd
import pymupdf
import os
```

```python
# Mamiferos

species_list = []
descriptions_list = []
for species in os.listdir('mamiferos'):
    for file in os.listdir(f'mamiferos/{species}'):
        if file.endswith('.pdf'):
            doc = pymupdf.open(f'mamiferos/{species}/{file}')
            text = ''
            for page in doc:
                text += page.get_text()
            try:
                text = text.split('DESCRIPCIÓN')[1]
                text = text.split('DISTRIBUCIÓN')[0]
                text = text.split('ULTRASONIDOS')[0]
            except:
                print(f'Cannot extract description for {species}')
            text = text.replace('-\n', '').replace('\n', ' ').strip()
            species_list.append(species)
            descriptions_list.append(text)
mamiferos_df = pd.DataFrame({'Nombre aceptado': species_list, 'Descripción': descriptions_list})
mamiferos_df['Fuente'] = 'Inventario IEET - Mamíferos'
mamiferos_df['Grupo taxonómico'] = 'Mamíferos'
mamiferos_df.to_excel('Descripciones_Mamíferos_IEET.xlsx', index=False)
mamiferos_df.head()
```

```python
# Peces

species_list = []
descriptions_list = []
for species in os.listdir('peces'):
    for file in os.listdir(f'peces/{species}'):
        if file.endswith('.pdf'):
            doc = pymupdf.open(f'peces/{species}/{file}')
            text = ''
            for page in doc:
                text += page.get_text()
            try:
                text = text.split('DESCRIPCIÓN')[1]
                text = text.split('POBLACIÓN')[0]
            except:
                print(f'Cannot extract description for {species}')
            text = text.replace('-\n', '').replace('\n', ' ').replace('BANCO DE DATOS DE LA NATURALEZA', '').replace('Peces Continentales de España', '')
            old_text = ''
            while old_text != text:
                old_text = text
                text = text.replace('  ', ' ')
            text = text.strip()
            species_list.append(species)
            descriptions_list.append(text)
peces_df = pd.DataFrame({'Nombre aceptado': species_list, 'Descripción': descriptions_list})
peces_df['Fuente'] = 'Inventario IEET - Peces'
peces_df['Grupo taxonómico'] = 'Peces'
peces_df.to_excel('Descripciones_Peces_IEET.xlsx', index=False)
peces_df.head()
```
